```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from pylab import *
```

```python
iq=np.array([104.3,104.2,103.8,103.8,103.7,103.4,103.4,103.2,103.1,102.9,102.8,102.8,102.8,102.4,102.3,101.9,101.9,101.8,101.7,101.6,101.5,101.4,101.2,101.1,101,100.7,100.5,100.4,100.2,100,99.9,99.7,99.5,99.4,99.3,99,98.7,98.4,98.4,98,97.7,97.5,97.4,96.5,95.7,95.7,95.6,95.5,95.3,94.2])
vaccinations=np.array([72.3,64.3,45.1,75.3,58.5,63.3,49,52.9,69.4,55.3,52.7,65.4,52.3,41.2,53.4,61.2,63.6,49.5,46.8,59.9,65,40.8,60.3,51.5,48,62.6,52.8,60.2,50.7,50.9,61.8,64.2,66.7,51.5,47.2,51.2,45.9,56.9,46.2,45.1,44.2,45.7,52.6,53,42.3,64.9,71.1,64.2,41.3,38.6])
```

```python


reg = linear_model.LinearRegression(fit_intercept=True)

# scikit-leearn linear regressions mode requires thast x be of shape (20,1)
# and not (20,)

labels=iq.reshape(iq.shape[0],1)
features=vaccinations

# fit means calculate the coeffiecient and y-intercept

reg.fit(labels,features)
```

```python
print('Coefficients: \n', reg.coef_)
print('Intercept: \n', reg.intercept_)
print('Mean squared error: %.2f' % mean_squared_error(labels, features))
print('Coefficient of determination: %.2f' % r2_score(labels, features))
```

```python
predictions = (reg.coef_[0] * labels) + reg.intercept_
```

```python
plt.plot(labels,predictions,color='b')
plt.plot(labels,features,color='g')
plt.title("Vaccination and IQ plot")
plt.ylabel("Vaccination Rates %")
plt.xlabel("IQ Numbers")
```
