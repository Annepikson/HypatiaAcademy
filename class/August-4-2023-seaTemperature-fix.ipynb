```python

#The data is here.  Use this to copy it to google colab

#!wget https://raw.githubusercontent.com/werowe/HypatiaAcademy/master/class/seaTemperature.csv



import re

file = open('/Users/walkerrowe/Documents/hypatia/sea/seaTemperature.csv')


# sections
s=[]

# one section

readings=[]
reading=[]

lines=file.readlines()

# the goal here is to to add each line to the line r

# so we read the first like which says '#-----' and keep reading until we see #-----'
# again
# we add each line that w read to r using r.append(stl(count))
# here add r to s using s.append(r)
# then 
# erase r by writing r=[]

count = 0
start=False
 
for l in lines:
    
    x = re.search("#--.", l)
  
    if x and not start:
        start=True
    else:
        # when hit next line add array to readings then erase array for the next section
        if x:
            # we hit another hash tag so add array reading to readings
            # then erase array reading
            readings.append(reading)
            reading=[]
            
          
    # keep adding lines to r until you hit another so
    
    # add line
    if not re.search("#--.", l):
        reading.append(l.split(","))
    
    
```
