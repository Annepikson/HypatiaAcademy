```python
import pandas as pd

file="https://raw.githubusercontent.com/werowe/HypatiaAcademy/master/pandas/DataScience_salaries_2024.csv"
df = pd.read_csv(file)
df
```

```python

# Filter rows where company_location is "US"
us_companies = df.loc[df['company_location'] == 'US']

# Display the filtered DataFrame
print(us_companies)
```

```python

# Filter rows where company_location is "FR" (France), and select specific columns
france_data = df.loc[df['company_location'] == 'FR', ['salary', 'job_title', 'experience_level']]

# Display the filtered DataFrame
print(france_data)
```

```python


# Select the first 3 rows and the first 6 columns using .iloc
select = df.iloc[:3, :6]

# Display the selected subset
print(select)
```
